```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
import numpy as np
df['target'] = np.where(df['target'] =='spam',1,0)
df.head()
```

```python
spam = []
ham = []
spam_class = df[df['target']==1]['desc']
ham_class = df[df['target']==0]['desc']
```

```python
ham_class.head()
```

```python
spam_class.head()
```

Comme on le voit, on créer deux autres dataframe qui détiennent chacune les Spams et Hams séparement.

```python
#on va créer ces deux fonctions qui extraient tous les mots correctes des deux Df
def extract_ham(ham_class):
    global ham
    words = [word.lower() for word in word_tokenize(ham_class) if word.lower() not in stopwords.words("english") and word.lower().isalpha()]
    ham = ham + words
def extract_spam(spam_class):
    global spam
    words = [word.lower() for word in word_tokenize(spam_class) if word.lower() not in stopwords.words("english") and word.lower().isalpha()]
    spam = spam + words
```

```python
spam_class.apply(extract_spam)
```

```python
ham_class.apply(extract_ham)
```

## Nettoyage de données

```python
import string
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer("english")

def cleanText(message):
    
    message = message.translate(str.maketrans('', '', string.punctuation))
    words = [stemmer.stem(word) for word in message.split() if word.lower() not in stopwords.words("english")]
    
    return " ".join(words)
```

```python
df["desc"] = df["desc"].apply(cleanText)
df.head(n = 10)
```

Comme on peut le voir, le stemmer a pour but transformer   
connection, connections, connective, connected et connecting en connect

## Model

On va créer un model de machine learning qui va nous premettre de prédire si un futur mail sera un Spam ou un Ham

```python
X = df.desc
y = df.target
```

```python
#on installe sklearn qui est une librairie qui va nous etre utile
!pip install sklearn
```

```python
#cet outil va nous aider à mieux prédire puisqu'il va diviser les données en deux sets: Training Set and Validation Set
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score

#cv va nous convertir nos données en matrice de nombre de Tokens
cv = CountVectorizer()

X_train = cv.fit_transform(X_train)
```

Essayons avec plusieurs models différents:

```python
from sklearn.linear_model import LogisticRegression

#lr est donc un model
lr = LogisticRegression(max_iter=10000)

lr.fit(X_train,y_train)
pred_1 = lr.predict(cv.transform(X_test))
score_1 = accuracy_score(y_test,pred_1)
score_1
```

```python
from sklearn.naive_bayes import MultinomialNB

#nb est donc un model
nb = MultinomialNB()

nb.fit(X_train,y_train)
pred_2 = nb.predict(cv.transform(X_test))
score_2 = accuracy_score(y_test,pred_2)
score_2
```

```python
from sklearn.svm import SVC

#svm est donc un model
svm = SVC()

svm.fit(X_train,y_train)
pred_3 = svm.predict(cv.transform(X_test))
score_3 = accuracy_score(y_test,pred_3)
score_3
```

## Conclusion

Entre les 3 modèles qu'on utilise, on voit que le meilleur est Multinomial Naive Bayes.
